```python
csv_fn = "control_data.csv"

path = os.getcwd()
path = path[:-4]

source_path = path + 'measurements/sbrd_output/dima'
destination_path = path + 'measurements/sbrd_output/antibiotics/analysis'

out_name = out_name_root + csv_fn

csv_files = [os.path.join(source_path, fn) for fn in os.listdir(source_path) if fn.endswith("comparison.csv") and fn != csv_fn]

comparisons = []

for csv_file in csv_files:
    df = pd.read_csv(csv_file, index_col = 0, header = 0)

    delta = df["D_left"] - df["D_right"]
    delta = delta.to_list()
    delta_sbrd = df["D_left_SBRD"] - df["D_right_SBRD"]
    delta_sbrd = delta_sbrd.to_list()
    left_is_new_pole = df['left_is_new'].to_list()

    check = []

    for element in left_is_new_pole:
        if type(element) == bool:
            check.append(str(element))
        else:
            check.append(element)
    
    test = []
    
    for i in range(len(left_is_new_pole)):
        
        if left_is_new_pole[i] == 'None':
            test.append('None')
        else:
            value = (((check[i] == "True") & (delta[i] > 0)) | ((check[i] == "False") & (delta[i] < 0)))
            test.append(value)
        
    test_sbrd = []
    
    for i in range(len(left_is_new_pole)):
        
        if left_is_new_pole[i] == 'None':
            test_sbrd.append('None')
        else:
            value = (((check[i] == "True") & (delta_sbrd[i] > 0)) | ((check[i] == "False") & (delta_sbrd[i] < 0)))
            test_sbrd.append(value)
    
    comparison = pd.DataFrame(
        {
            "name": csv_file,
            "left_is_new_pole": df["left_is_new"],
            "D_left": df["D_left"],
            "D_center": df["D_center"],
            "D_right": df["D_right"],
            "delta": delta,
            "D_left_SBRD": df["D_left_SBRD"],
            "D_center_SBRD": df["D_center_SBRD"],
            "D_right_SBRD": df["D_right_SBRD"],
            "delta_SBRD": delta_sbrd,
            "test": test,
            "test_SBRD": test_sbrd
        }
    )
    
    comparisons.append(comparison)    
    

control_data = pd.concat(comparisons, ignore_index=True)

control_data.to_csv(out_name)
```

```python
control_center = control_data['D_center_SBRD']
control_poles = (control_data['D_left_SBRD'] + control_data['D_right_SBRD'])/2
control_ratio = control_poles/control_center

control_center_mean = np.mean(control_center)
control_center_stdev = np.std(control_center)
control_poles_mean = np.mean(control_poles)
control_poles_stdev = np.std(control_poles)
control_ratio_mean = np.mean(control_ratio)
control_ratio_stdev = np.std(control_ratio)


erythromycin_center = erythromycin_data['D_center_SBRD']
erythromycin_poles = (erythromycin_data['D_left_SBRD'] + erythromycin_data['D_right_SBRD'])/2
erythromycin_ratio = erythromycin_poles/erythromycin_center

erythromycin_center_mean = np.mean(erythromycin_center)
erythromycin_center_stdev = np.std(erythromycin_center)
erythromycin_poles_mean = np.mean(erythromycin_poles)
erythromycin_poles_stdev = np.std(erythromycin_poles)
erythromycin_ratio_mean = np.mean(erythromycin_ratio)
erythromycin_ratio_stdev = np.std(erythromycin_ratio)


rifampicin_center = rifampicin_data['D_center_SBRD']
rifampicin_poles = (rifampicin_data['D_left_SBRD'] + rifampicin_data['D_right_SBRD'])/2
rifampicin_ratio = rifampicin_poles/rifampicin_center

rifampicin_center_mean = np.mean(rifampicin_center)
rifampicin_center_stdev = np.std(rifampicin_center)
rifampicin_poles_mean = np.mean(rifampicin_poles)
rifampicin_poles_stdev = np.std(rifampicin_poles)
rifampicin_ratio_mean = np.mean(rifampicin_ratio)
rifampicin_ratio_stdev = np.std(rifampicin_ratio)
```
